##  All the above steps joined together to form one function


````python
import re
def answer_one():
    # Importing the data
    df1=pd.read_excel('Energy Indicators.xls',header=[1,2],skiprows=15,usecols=[2,3,4,5],skip_footer=38)

    # you should change the column labels so that the columns are:['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    df1.rename(columns={'Renewable Electricity Production':'% Renewable'}, inplace=True)
    
    # copying the name(index) to a new column called country
    df1['Country']=df1.index

    # Data Cleaning
    
    # Cleaning numbers from country name  (sucessfull)
    pattern = r'([a-zA-Z\s]+)([0-9]+)'   # our country name is combination of  capital and small alphabets,seprated by space and ending withnumber
    result= df1['Country'].str.split(pattern, expand=True)
    # now  Joining column 0 and column 1 to get color a column of ful name
    result['Colors'] = result[[0,1,]].apply(lambda x: ''.join(x[x.notnull()]), axis = 1)
    df1['Country1']=result['Colors']  # putting the  content of color (ful name) in new column  country1
    df1['Country']=0
    
    # Replace all the bracket and contents within bracket with ' '
    df1['Country1'] = df1['Country1'].apply(lambda x: re.sub(r'\(.*\)', '', x))
    
    # strip returns the string after stripping whitespace.
    # 'Bolivia ' -->'Bolivia'
    df1['Country2'] = df1['Country1'].map(lambda x: x.strip())
    
    
    
    # Now replacing contents
    # replacing name of  selected countries
    df1=df1.replace({"Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland19": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"})
    df1
    
    
    #  setting Column country2 as our index
    df1=df1.set_index('Country2')
    del df1['Country1']
    del df1['Country']
    
    energy=df1
    
    return energy

answer_one()
````
