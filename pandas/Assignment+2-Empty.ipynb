```python
import pandas as pd
df=pd.read_csv('olympics.csv',skiprows=1,skipfooter=0)

i=0
columns=df.columns
for col in columns:
    if (col[:1]=='№'):
        df.rename(columns={col: '# '+col[2:]},inplace=True)
    elif (col[:2]=='01'):
        df.rename(columns={col: 'Gold'+col[5:]},inplace=True)    
    elif (col[:2]=='02'):
        df.rename(columns={col: 'Silver'+col[5:]},inplace=True)
    elif (col[:2]=='03'):
        df.rename(columns={col: 'Bronze'+col[5:]},inplace=True)
           
df.set_index('Unnamed: 0',inplace=True)    # making country name our index
# cleaning the index 
name_id=df.index.str.split('(')            
index=name_id.str[0].str[0:-1]
id=name_id.str[1].str[0:3]
df['ID']=id
df.set_index(index,inplace=True)
df
```
