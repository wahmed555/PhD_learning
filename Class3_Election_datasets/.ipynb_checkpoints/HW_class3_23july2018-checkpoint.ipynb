```python
parties=pd.read_csv('parties.csv')
# loading file "parties.csv"  containing the perfect names without any mistake list 
```

```python
def resolve_party(x):
    max=0
    match=''

    for i in parties['Name of Political Party']:
        r=fuzz.ratio(x,i)
        if r > max:
            max=r
            match=i
    return match
```

#### Find out the best match for a party name, resolving spelling mistakes

```python
extracted_parties=df.Party.apply(lambda x: resolve_party(x))
```
