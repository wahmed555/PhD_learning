# adding after the class fuzy wuzzy use

```python
parties=pd.read_csv('parties.csv')
# loading file "parties.csv"  containing the perfect names without any mistake list 
```

```python
def resolve_party(x):
    max=0
    match=''

    for i in parties['Name of Political Party']:
        r=fuzz.ratio(x,i)
        if r > max:
            max=r
            match=i
    return match
```

Find out the best match for a party name, resolving spelling mistakesÂ¶

```python
extracted_parties=df1.Party.apply(lambda x: resolve_party(x))
```

correcting Typing Mistakes in the name of Parties by replacing with correct Party names

```python
df1['Party']=extracted_parties
```

### Find number of candidates per partyÂ¶

```python
fight4_seats=df1.groupby(['Party'])['Seat'].count()
```

### Parties Fighting for largest number of seatsÂ¶

```python
Big_parties=fight4_seats.nlargest(12)
```
